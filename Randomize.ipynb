```python
import random
from math import *
```

```python
random.randint(2,8) ## randint between 2 values
```

```python
random.randrange(10,100,15) ## randint from range
```

```python
arr = [tanh(i)/(1+exp2(-i)) for i in range(-10,10)]
random.choice(arr)      ## random choice
```

```python
import matplotlib.pyplot as plt
```

```python
plt.grid()
x = [i for i in range(-10,10)]
plt.hlines(0,-10,10, 'black','dotted')
plt.vlines(0,-5,5,'black', 'dotted')
plt.ylim(arr[0]-1,arr[-1]+1)
plt.plot(x,arr, 'r')
y = random.choice(arr)
plt.plot(x,[y for i in range(-10,10)], 'c-+')
```

```python
print(bin(random.getrandbits(8))) # return an integer within the specified size in bits
```

```python
random.choices(arr,k=5) ## random selection, with duplicates, no-limit k
```

```python
random.sample(arr, k = 2) ## number of random elements from a sequence
                        ## k <= len(arr), cannot be negative
```

```python
random.uniform(0.5,1.5) # random float between two numbers
```

```python
random.triangular(0,100,0.1) # triangular distribution
```
